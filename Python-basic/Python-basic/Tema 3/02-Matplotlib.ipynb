# Utilizar gráficos en Python

Una de las bibliotecas gráficas más utilizadas en python es **MatPlotLib**. Es una biblioteca básica, a pesar de haver otras con más funcionalidades


```python
# Iniciar siempre con código para que los gráficos salen bien en jupyter.
%matplotlib inline
```

```python
# Llamar la biblioteca pyplot de matplotlib
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
```

```python
plt.plot(x)
```

Se observa que con `R` al llamar plot con sólo una variable, plotaba puntos al invés de lineas como hace aqui con python.

Se puede modificar tambíen las características de la gráfica

Ejemplo:

```python
plt.plot(x)
plt.xlabel("Eje de las abcisas")
plt.ylabel("Eje de las ordenadas")
plt.show()# para actualizar el gráfico se ejecuta la función show
```

**Nota importante:** Si suministra sólo un dato, como en el ejemplo anterior, plot pintará el gráfico con los datos de las ordenadas (eje y) como ocurre con `R`.

```python
#suministrando los valores de x e y
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y)
```

Si uno de los vectores de datos tiene dimenciones distintas, provocará un error al intentar plotar.

## Para plotar puntos
En python plota en linea en modo estandar. Para plotar en puntos como en `R` se hace asi:

```python
plt.plot(x,y,"ro")
```

donde **r** significa el color rojo y **o** significa puntos.

Para modificar el tamaño de de los ejes se utiliza la función **axis** de la forma axis([$x_{inicial},x_{final},y_{inicial},y_{final}$]).

```python
plt.plot(x,y,"ro")
plt.axis([0,6,0,20])
plt.show()
#Al tener más de una función, hay que llamar siempre la función show para plotar.
```

## Multiples gráficos en la misma figura

```python
#Declarando un función
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
#Declarar dos vectores para el ejemplo
x1 = np.arange(0, 5, 0.1)
x2 = np.arange(0, 5, 0.2)
```

```python
plt.figure(1)

plt.subplot(211)
plt.plot(x1, f(x1), "ro", x2, f(x2), "k")

plt.subplot(212)
plt.plot(x1, f(x1), "g--")
plt.show
```

Se puede hacer lo mismo que el anterior donde el **subplot** puede ir separado por comas. Recordando que el primero representa el número de filas, el segundo el número de columnas y el último el index de la figura que vas a plotar.

```python
plt.figure(1)

plt.subplot(2,1,1)
plt.plot(x1, f(x1), "ro", x2, f(x2), "k")

plt.subplot(2,1,2)
plt.plot(x1, f(x1), "g--")
plt.show
```

Los ejemplos anteriores son dos plot en la misma figura separados por zonas. Abajo muestra la diferencia:

```python
plt.figure(1)

plt.subplot(2,1,1)
plt.title("Gráficos de la figura 1")
plt.plot([1,2,3])
plt.subplot(2,1,2)
plt.plot([4,5,6])

#Plotando el ultimo dato en una figura distinta
plt.figure(2)
plt.plot([4,5,6])
plt.title("Gráfico de Figura Completa")
plt.show
```

**Nota importante:** Si no cierra las figuras, a medida que se va creando figuras se queda almacenado en memoria.

## Textos en los gráficos

```python
mu = 100 #media
sigma = 20 #desviación típica
#Generar datos aleatorios
x = mu + sigma * np.random.randn(10000)
```

```python
n, bins, patches = plt.hist(x, 50, normed=1, facecolor="g", alpha=0.6)
plt.show()
```

El histograma anterior esta plotado de forma normalizada(datos entre 0 y 1), con el parámetro **normed=1**, con un color verde, con el parámetro **facecolor="g"** y con un nivel de transparencia con el parámetro **alpha=0.6**.

Para poner etiquetas en el eje X y en el eje Y.

```python
n, bins, patches = plt.hist(x, 50, normed=1, facecolor="g", alpha=0.6)
#Poner etiquetas y título
plt.xlabel("Cociente intelectual")
plt.ylabel("Probabilidad")
plt.title("Histograma de Cociente Intelectual")
plt.show()
```

Para poner texto dentro del gráfico

```python
n, bins, patches = plt.hist(x, 50, normed=1, facecolor="g", alpha=0.6)
#Poner etiquetas y título
plt.xlabel("Cociente intelectual")
plt.ylabel("Probabilidad")
plt.title("Histograma de Cociente Intelectual")

#Poner un texto dentro del gráfico con text(x,y,"texto") siguiendo esa sintaxis.
plt.text(40, 0.015, r'$\mu = 100,\  \sigma = 20$')
plt.show()
```

Además del texto se puede tambien activar la parrilla con la función **grid=True**

```python
n, bins, patches = plt.hist(x, 50, normed=1, facecolor="g", alpha=0.6)
#Poner etiquetas y título
plt.xlabel("Cociente intelectual")
plt.ylabel("Probabilidad")
plt.title("Histograma de Cociente Intelectual")

#Poner un texto dentro del gráfico con text(x,y,"texto") siguiendo esa sintaxis.
plt.text(40, 0.015, r'$\mu = 100,\  \sigma = 20$')

# Activando la parrilla de la gráfica
plt.grid(True)
plt.show()
```

Para ajustar los ejes de la gráfica para que quede más bonito

```python
n, bins, patches = plt.hist(x, 50, normed=1, facecolor="g", alpha=0.6)
#Poner etiquetas y título
plt.xlabel("Cociente intelectual")
plt.ylabel("Probabilidad")
plt.title("Histograma de Cociente Intelectual")

#Poner un texto dentro del gráfico con text(x,y,"texto") siguiendo esa sintaxis.
plt.text(40, 0.015, r'$\mu = 100,\  \sigma = 20$')

# Activando la parrilla de la gráfica
plt.grid(True)

#Ajustando los ejes de la gráfica
plt.axis([20, 180, 0, 0.0210])#Poner siempre dentro de una lista
plt.show()
```

## Decorativos en los gráficos

Hay momentos que se desea poner determinados símbolos en la gráfica para dar más énfasis en la información.

Ejemplo:


```python
#Figura con ancho de 10' y un alto de 6' con una calidad gráfica de 90 dpi.
plt.figure(figsize = (10,6), dpi = 90)
plt.subplot(1,1,1)

x = np.arange(0, 2*np.pi, 0.01)
y = np.cos(x)
plt.plot(x, y, lw = 2.0)
```

La gráfica anterior plota el coseno con un intervalor de $[0\ a\ 2\pi]$

Si quiere almentar el intervalo basta modificalo en:

```python
#Figura con ancho de 10' y un alto de 6' con una calidad gráfica de 90 dpi.
plt.figure(figsize = (10,6), dpi = 90)
plt.subplot(1,1,1)

x = np.arange(0, 10*np.pi, 0.01)
y = np.cos(x)
plt.plot(x, y, lw = 2.0)
```

La gráfica anterior plota el coseno con un intervalor de $[0\ a\ 10\pi]$.

Ahora si quieres marcar los máximos que cumpla con el intervalo de $2\pi$.

```python
#Figura con ancho de 10' y un alto de 6' con una calidad gráfica de 90 dpi.
plt.figure(figsize = (10,6), dpi = 90)
plt.subplot(1,1,1)

x = np.arange(0, 10*np.pi, 0.01)
y = np.cos(x)
plt.plot(x, y, lw = 2.0)

#Poner una marca en el punto deseado
plt.annotate('Máximo Local', xy = (4*np.pi,1), xytext = (15, 1.5),
            arrowprops = dict(facecolor = "black", shrink = 0.05))

#Modificar el tamaño del lienzo de la gráfica. Sólo el eje y.
plt.ylim(-2, 2)

plt.show()
```

### Cambios de ejes y escalas

```python
mu = 0.5
sd = 0.3
y = mu+sd*np.random.randn(1000)
y = y[(y>0) & (y<1)]#Aplicando filtro para valores de 0 a 1
y.sort() #Ordenando los valores de menor a mayor
x = np.arange(len(y))
```

Plotar los datos generados en la línea anterior

```python
from matplotlib.ticker import NullFormatter

plt.figure(figsize = (10,8))#Fija el lienzo en 10 de alto y 8 de ancho
plt.subplot(2,2,1)
plt.plot(x,y)
plt.yscale("linear")
plt.xscale("linear")
plt.title("Escala Lineal")
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(x,y)
plt.yscale("log")
plt.title("Escala Logaritmica")
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(x,y - y.mean())
plt.yscale("symlog", linthreshy=0.01)#Simetrico logarítmo con intervalos de 0.01
plt.title("Escala Log Simétrico")

plt.grid(True)

plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale("logit")#Escala Logística
plt.title("Escala Logística")
plt.gca().yaxis.set_minor_formatter(NullFormatter())#Eliminar números muy pequeños
plt.grid(True)

#Ajustar las posiciones de los gráficos
plt.subplots_adjust(top = 0.92, bottom = 0.08, left = 0.1, right = 0.95, hspace = 0.35, wspace = 0.35)
plt.show()
```
