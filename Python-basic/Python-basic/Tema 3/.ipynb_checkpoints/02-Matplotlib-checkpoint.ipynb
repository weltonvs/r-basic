# Utilizar gráficos en Phyton

Una de las bibliotecas gráficas más utilizadas en phyton es **MatPlotLib**. Es una biblioteca básica, a pesar de haver otras con más funcionalidades


```python
# Iniciar siempre con código para que los gráficos salen bien en jupyter.
%matplotlib inline
```

```python
# Llamar la biblioteca pyplot de matplotlib
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
```

```python
plt.plot(x)
```

Se observa que con `R` al llamar plot con sólo una variable, plotaba puntos al invés de lineas como hace aqui con phyton.

Se puede modificar tambíen las características de la gráfica

Ejemplo:

```python
plt.plot(x)
plt.xlabel("Eje de las abcisas")
plt.ylabel("Eje de las ordenadas")
plt.show()# para actualizar el gráfico se ejecuta la función show
```

**Nota importante:** Si suministra sólo un dato, como en el ejemplo anterior, plot pintará el gráfico con los datos de las ordenadas (eje y) como ocurre con `R`.

```python
#suministrando los valores de x e y
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y)
```

Si uno de los vectores de datos tiene dimenciones distintas, provocará un error al intentar plotar.

## Para plotar puntos
En phyton plota en linea en modo estandar. Para plotar en puntos como en `R` se hace asi:

```python
plt.plot(x,y,"ro")
```

donde **r** significa el color rojo y **o** significa puntos.

Para modificar el tamaño de de los ejes se utiliza la función **axis** de la forma axis([$x_{inicial},x_{final},y_{inicial},y_{final}$]).

```python
plt.plot(x,y,"ro")
plt.axis([0,6,0,20])
plt.show()
#Al tener más de una función, hay que llamar siempre la función show para plotar.
```

## Varias funciones en una misma gráfica


```python
#utilizar numpy para generar secuencias de datos
import numpy as np
```

```python
datos = np.arange(0.0, 10.0, 0.2)
datos
```

```python
plt.plot(datos, datos, "r--")
plt.show()
```

```python
plt.plot(datos, datos, "r--", datos, datos**2,"bs")
plt.show()
```

```python
plt.plot(datos, datos, "r--", datos, datos**2,"bs", datos, datos**3, "g^")
plt.show()
```

Como se puede observar se puede en la misma línea de plot ir definiendo varias gráficas

```python
#Para cambiar el grosor de la línea
plt.plot(x, y, linewidth = 5.0)
plt.show()
```

## Efecto del antialiasing

```python
line, = plt.plot(x, y, "-")
line.set_antialiased(False)
```

## Parámetros adicionales
Muy utilizado para controlar varios plots dentro de una misma gráfica

```python
lines = plt.plot(datos,datos, datos, datos**2)
plt.setp(lines, color = "r", linewidth = 2.0)#Sintaxis de argumentos
plt.show()
```

Aqui se cambia los parametros grobales de la gráfica.

```python
lines = plt.plot(datos,datos, datos, datos**2)
plt.setp(lines, "color", "r", "linewidth", 2.0)#Sintaxis de parámetros directamente
plt.show()
```

Ambos resultados son iguales

Otros ejemplos interesantes:

```python
plt.plot(x,y,marker = "+", linestyle = ":")
plt.show()
```

Para saber todos los parámetros:

```python
plt.setp(lines)
```
